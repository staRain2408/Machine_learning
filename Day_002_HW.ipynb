了解機器學習適合應用的領域與範疇。

## [作業重點]
透過瞭解機器學習的運作方式，找出非常適用機器學習應用的領域及其原因

## 作業
請找出一個你認為機器學習可以發光發熱的領域並闡述原因? 

領域:都市災防預警系統

應用設想:當今全球已經有過半數的人口居住於都市之中，同時都市中的人口密度普遍高於其餘非都市區域。因此假若都市中相關自然或人為災害，將會連帶影響到較多的人口範圍。同時，在全球氣候變遷與極端氣候發生頻率逐漸提升的今日，都市中的災害應變能力成為影響民住居住品質的慣見要素之一。

透過於都市之中的各處部屬監測器，已蒐集即時資料，並搭配網際網路傳輸資訊至後台。配合背景機器學習處理分析資料型態，藉此推估災害發生的潛在區域與災害嚴重程度，藉此提前發布防災警訊並提前部屬防災物資。令該都市應變災害的能力有效提升。
